## First batch into a dictionary

This file will be used to take in the first batch on manual summaries by scene (popular shakespeare work), and put them in a dictionary with key the scene (matching the key of the text file) and value is a string which is the summary

```python
#Importing text
raw_text_location = "../data/summaries/"

#Importing raw text into strings
def read_raw_text(file_name):
    with open(raw_text_location+file_name, "r") as f:
        contents = f.read()
    return contents
```

```python
#Read in summary.txt files
summary_file_names = [
    "Hamlet",
    "Julius_Caesar",
    "MacBeth",
    "Othello",
    "Romeo_and_Juliet",
    "Twelfth_Night"
]

#a dictionary where key is the file name and the value is the summary itself
summary_dict = {e: read_raw_text(e+".txt") for e in summary_file_names}
```

```python
#Bringing in text dictionary
%store -r all_text_dict
```

## MacBeth

```python
#Looking at the appropriate cells
scenes_MacBeth = []
for k,v in all_text_dict.items():
    if "MacBeth" in k:
        scenes_MacBeth.append(k)
```

```python
#Making MacBeth Summary Dict (manual)
macbeth_lines = summary_dict["MacBeth"].split("\n")
macbeth_summary_dict = {}
for i in range(0,len(macbeth_lines)-2,2):
    scene_summary = macbeth_lines[i+1]
    
    macbeth_summary_dict[scenes_MacBeth[i//2]] = scene_summary
```
