## Second Batch batch into a dictionary

This file is loading the documents for summaries by scene for Ben Johnon plays (currently only Every Man in his Humour)

```python
#Importing text
raw_text_location = "../data/summaries/"

#Importing raw text into strings
def read_raw_text(file_name):
    with open(raw_text_location+file_name, "r") as f:
        contents = f.read()
    return contents
```

```python
#Read in summary.txt files
summary_file_names = [
    "Every_Man_in_His_Humour"
]

#a dictionary where key is the file name and the value is the summary itself
summary_dict = {e: read_raw_text(e+".txt") for e in summary_file_names}
```

```python
summary_dict.keys()
```

```python
#Given a summary text in a specific format, this breaks out the text into summaries
def summary_breakout_scenes(raw_text_file, story_name):
    
    summary_scene_dict={}

    scene_name_and_text = raw_text_file.split("\n\n")
    for e in scene_name_and_text:
        scene_split = e.split("\n")
        #Make the scene name match the scene name of the raw text
        scene_name=story_name+"-"+scene_split[0]

        summary_scene_dict[scene_name]=scene_split[1]
    
    return summary_scene_dict
```

```python
#Makes a dictionary for all the summaries in summary_dict

summaries_batch2={}

for summary_name,summary_contents in summary_dict.items():
    summaries_batch2.update(summary_breakout_scenes(summary_contents, summary_name))
```
