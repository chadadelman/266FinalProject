## Importing Raw data into the Notebook

```python
#Set up
raw_text_location = "../data/raw/Text/"
raw_summary_location = "../data/raw/Summaries/"


#Importing raw text into strings
def read_raw_text(file_name):
    with open(raw_text_location+file_name, "r") as f:
        contents = f.read()
    return contents

#Importing the raw summaries into strings
def read_raw_summaries(file_name):
    with open(raw_summary_location+file_name, "r") as f:
        contents = f.read()
    return contents
```
