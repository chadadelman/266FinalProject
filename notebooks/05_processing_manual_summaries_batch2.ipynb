## Third Batch batch into a dictionary

This file is loading the Shakespeare Sonnets

```python
#loading the text file
with open("../data/summaries/sonnets.txt", "r") as f:
    text = f.read()
```

```python
#breaking the text file into a dictionary
summaries_batch3={}

sonnets_text = text.split("\n\n")
for e in sonnets_text:
    sonnet_sum_split = e.split("\n")
    summaries_batch3[sonnet_sum_split[0]] = sonnet_sum_split[1]
```
