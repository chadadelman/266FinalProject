```python
#Importing text
raw_text_location = "../data/text/jonson/"

#Importing raw text into strings
def read_raw_text(file_name):
    with open(raw_text_location+file_name, "r") as f:
        contents = f.read()
    return contents
```

```python
text_file_names = [
    "Every_Man_in_His_Humour",
    "Every_Man_out_of_His_Humour",
    "The_Poetaster",
    "Sejanus_His_Fall",
    "Volpone_Or_The_Fox",
    "Epicoene_Or_The_Silent_Woman",
    "The_Alchemist",
    "Bartholomew_Fair",
    "The_Devil_is_an_Ass"
]

text_dict = {e: read_raw_text(e+"_wsplits.txt") for e in text_file_names}
```

```python
#Makes a function which breaks the text into scenes
#input is a dictianry of scene names and line that is the start of the scene, and the raw text
#output is a dictionary that has each scene title (as input), and the text
def breakout_scenes(raw_text):
    
    scene_list=[]
    remaining_text=raw_text
    split_text="*--split scene here--*"

    #if there is another split scene left
    while split_text in remaining_text:
        start_of_new_scene = remaining_text.find(split_text)
        
        #Lets only look at text after the new scene starts (not inculding split sentence)
        remaining_text = remaining_text[start_of_new_scene+22:]
        next_new_scene = remaining_text.find(split_text)

        #if last scene
        if next_new_scene ==-1:
            scene_list.append(remaining_text[: remaining_text.find("*** END OF THE PROJECT")])
        else:
            scene_list.append(remaining_text[: next_new_scene])


    return scene_list

        
```
