```python
with open("../data/sonnets/sonnets.txt", "r") as f:
    text = f.read()
```

```python
#creates the sonnets_dict_pre variable which has sonnets_dict_pre[Sonnet Number] = Sonnet Text
sonnets_dict_pre={}

text_adjusted = text.replace("\n\n\n","\n\n")
current_text=text_adjusted

while ("\n\n" in current_text):
    start_of_sonnet=current_text.find("\n\n")
    end_of_sonnet = (start_of_sonnet+1) + current_text[start_of_sonnet+1:].find("\n\n")

    current_number="Sonnet "+current_text[:start_of_sonnet]
    current_sonnet = current_text[start_of_sonnet+2:end_of_sonnet]

    sonnets_dict_pre[current_number] = current_sonnet
    
    #takes off sonnet that has been added to dict
    current_text=current_text[end_of_sonnet+2:]

#manual fix for last element
sonnets_dict_pre['Sonnet CLIV.']=text_adjusted[-594:]
```

```python
#Helper function for below which returns an arabic number (as a integer) when given a roman numeral (as a string)
def return_arabic_number(roman_numeral):
    number=0
    i=0
    while i < len(roman_numeral):
        if i+1 < len(roman_numeral): #If not the last digit
            next2=roman_numeral[i:i+2] 
            if next2 == "IV":
                number+=4
                i+=2
                continue
            elif next2 == "IX":
                number+=9
                i+=2
                continue
            elif next2 == "XL":
                number+=40
                i+=2
                continue
            elif next2 == "XC":
                number+=90
                i+=2
                continue

        next1=roman_numeral[i] 
        if next1=="I": number+=1
        elif next1=="V": number+=5
        elif next1=="X": number+=10
        elif next1=="L": number+=50
        elif next1=="C": number+=100

        i+=1

    return number
```

```python
#Make sonnet arabic numbers instead of roman numerals (done somewhat manually)
sonnets_dict={}

for k,v in sonnets_dict_pre.items():
    rom_num = k[7:-1]
    ar_num = str(return_arabic_number(rom_num))

    sonnets_dict["Sonnet "+str(ar_num)]=v


#print(*sonnets_dict_pre.keys(),sep="\n")
#print(*sonnets_dict.keys(),sep="\n")
```
