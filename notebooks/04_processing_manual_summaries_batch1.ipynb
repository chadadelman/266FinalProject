## First batch into a dictionary

This file will be used to take in the first batch on manual summaries by scene (popular shakespeare work), and put them in a dictionary with key the scene (matching the key of the text file) and value is a string which is the summary

```python
#Importing text
raw_text_location = "../data/summaries/"

#Importing raw text into strings
def read_raw_text(file_name):
    with open(raw_text_location+file_name, "r") as f:
        contents = f.read()
    return contents
```

```python
#Read in summary.txt files
summary_file_names = [
    "Hamlet",
    "Julius_Caesar",
    "MacBeth",
    "Othello",
    "Romeo_and_Juliet",
    "Twelfth_Night"
]

#a dictionary where key is the file name and the value is the summary itself
summary_dict = {e: read_raw_text(e+".txt") for e in summary_file_names}
```

```python
summary_dict.keys()
```

```python
#Given a summary text in a specific format, this breaks out the text into summaries
def summary_breakout_scenes(raw_text_file, story_name):
    
    summary_scene_dict={}

    scene_name_and_text = raw_text_file.split("\n\n")
    for e in scene_name_and_text:
        scene_split = e.split("\n")
        #Make the scene name match the scene name of the raw text
        scene_name=story_name+"-"+scene_split[0].replace(", ","-")[:-1]

        #Turn arabic number into roman numeral
        scene_name=scene_name.replace("1","I")
        scene_name=scene_name.replace("2","II")
        scene_name=scene_name.replace("3","III")
        scene_name=scene_name.replace("4","IV")
        scene_name=scene_name.replace("5","V")
        scene_name=scene_name.replace("6","VI")
        scene_name=scene_name.replace("7","VII")
        scene_name=scene_name.replace("8","VIII")
        scene_name=scene_name.replace("9","IX")
        scene_name=scene_name.replace("10","X")
        scene_name=scene_name.replace("11","XI")
        scene_name=scene_name.replace("12","XII")

        summary_scene_dict[scene_name]=scene_split[1]
    
    return summary_scene_dict
```

```python
#Makes a dictionary for all the summaries in summary_dict

summaries_batch1={}

for summary_name,summary_contents in summary_dict.items():
    summaries_batch1.update(summary_breakout_scenes(summary_contents, summary_name))
```
